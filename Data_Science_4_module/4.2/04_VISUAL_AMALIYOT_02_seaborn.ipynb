### 3-VAZIFA. Grafiklar nomini to'g'rilang (`T`, `Tm` va hokazo matnlarni o'zbek tilida chiqaring.)

```python
#JAVOBINGIZNI SHU YERGA YOZING
df.rename({'T': "o'rtacha yillik harorat", 
'TM': "Yillik o'rtacha maksimal harorat",
'Tm': "O'rtacha yillik minimal harorat",
'PP': "Yillik jami yomg'ir yoki qor yog'inlari",
'V': "Yillik o'rtacha shamol tezligi",
'RA': "Yomg'irli kunlar soni",
"SN": "Qorli kunlar soni",
'TS': "Bo'ronli kunlar soni",
'FG': "Tumanli kunlar soni",
'TN': "Tornado bilan kunlar soni",
'GR': "Do'l yog'adigan kunlar soni"}, inplace= True, axis=1)
df.head()
plt.figure(figsize=[16,6])
sns.lineplot(data = df)
plt.legend(loc = 1)
plt.show()
```

```python
#@title Kutilgan natija:
labels=["Yillik o'rtacha harorat",
        "Yillik o'rtacha maksimal harorat",
        "Yillik o'rtacha minimal harorat",        
        "Yillik o'rtacha shamol tezligi",
        "Yomg'irli kunlar soni",
        "Bo'ronli kunlar soni",
        "Tumanli kunlar soni",
        "To'fonli kunlar soni",
        "Do'l yoqqan kunlar"
        ]

plt.figure(figsize=(16,6))
sns.lineplot(data=df, linewidth=2.0)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.legend(labels, loc=1)
plt.show()
```

### 5-VAZIFA. Yillik o'rtacha shamol tezligini ustunli grafik ko'rinishida chiqaring.

```python
#JAVOBINGIZNI SHU YERGA YOZING
plt.figure(figsize = [16,6])
sns.barplot(data = df[["Yillik o'rtacha shamol tezligi"]].T)
plt.title("Toshkent shahri uchun ob-havo ko'rsatgichlari")
plt.xticks(rotation = 90)
plt.xlabel("Yillar")
plt.ylabel("O'racha shamol tezligi (km/soat)")
plt.show()
```

```python
df.head()
```

```python
#@title Kutilgan natija:
plt.figure(figsize=(16,6))
sns.barplot(x=df.index, y=df['V'])
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.ylabel("O'rtacha shamol tezligi (km/soat)")
plt.xlabel("Yillar")
plt.xticks(rotation=90)
plt.show()
```

### 6-VAZIFA. Yomg'rili, qorli, bo'ronli, tumanli va do'l yoqqan kunlar sonini heatmap ko'rinishida chiqaring.

```python
#JAVOBINGIZNI SHU YERGA YOZING
plt.figure(figsize = [20,3])
mask = df[["Yomg'irli kunlar soni", "Qorli kunlar soni", "Bo'ronli kunlar soni", "Tumanli kunlar soni", "Do'l yog'adigan kunlar soni"]].T
sns.heatmap(data = mask, cmap = "YlGnBu", linewidth = 3, square = True )
plt.title("Toshkent shahar uchun ob-havo ko'rsatgichlari")
plt.xticks(rotation = 90)
plt.show()
```

```python
df.head()
```

```python
#@title Kutilgan natija:
plt.figure(figsize=(30,5))
sns.heatmap(data=df[['RA','SN','TS','FG','GR']].T, cmap="YlGnBu", linewidths=1, square=True)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
yticks_labels = ["Yomg'ir","Qor","Bo'ron","Tuman","Do'l"]
plt.yticks(np.arange(5)+0.5, labels=yticks_labels)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```
